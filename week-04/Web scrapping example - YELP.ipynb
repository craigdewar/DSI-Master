```python

data = soup.find_all("div", {"class" : "biz-listing-large"})
url = "https://www.yelp.com/search?find_desc=Restaurants&find_loc={}&start={}"
cities = ["oklahoma city", "atlanta", "new york", "miami", "nashville"]
df = pd.DataFrame()

for city in cities:
    for i in range(0,20,10):
        LINK = url.format(city, i)
        r = requests.get(LINK)
        soup = BeautifulSoup(r.content)
        data = soup.find_all("div", {"class" : "biz-listing-large"})

        for element in range(len(data)):
            Name = data[element].find("a", {"class": "biz-name js-analytics-click"}).get_text().strip()
        #     Ratings = data[element].find("div", {"class": "i-stars i-stars--regular-5 rating-large"}).title
            Phone = data[element].find("span", {"class": "biz-phone"}).get_text().strip()
            Address = data[element].find("address")
            if Address != None:
                Address2 = Address.get_text().strip()
                df = df.append({"Name" : Name, "Phone": Phone, "Address": Address2, "City": city}, ignore_index=True)

df
```
